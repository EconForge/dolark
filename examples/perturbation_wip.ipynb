```python
import dolark
```

```python
hmodel = dolark.HModel("ayiagari.yaml")    # no aggregate state
# hmodel = dolark.HModel("prototype.yaml") # with aggregate state
```

```python
from dolark.equilibrium import find_steady_state
from dolark.perturbation import FirstOrderModel
from dolark.perturbation import perturb, get_derivatives
from dolark.dolo_improvements import pack, unpack
from dolo.algos.perturbation import approximate_1st_order
```

```python
eq = find_steady_state(hmodel)
```

```python
peq = perturb(hmodel, eq)
```

```python
eq.hmodel = eq.aggmodel
```

```python
sim = peq.simulate(T=200)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.subplot(211)
plt.plot( sim.sel(V="z"))
plt.subplot(212)
plt.plot( sim.sel(V="K"))
```
